```python
import torch
```

```python
#1.创建一个未初始化的张量
x = torch.empty(5,3)
print("Empty Tensor:",x)
```

```python
"Empty Tensor:",x
```

```python
#2.创建一个随机初始化的张量
x = torch.rand(5,3)
print("Random Tensor:\n",x)
```

```python
x = torch.randn(5,3)#根据正态分布进行随机初始化
print("random tensor with normal distribution:\n",x)
```

```python
#3. 创建一个全零张量，并指定数据类型为long
x = torch.zeros(5,3,dtype=torch.long)
print("Zero Tensor:\n",x)
```

```python
#4.直接从数据创建张量
x = torch.tensor([5.5,3])
print("Directly create tensor from data:\n",x)
```

```python
x = torch.ones(3,4)
y = torch.ones_like(x)#根据x的形状创建一个全1张量
print("Ones like x:\n",y)
```

```python
x = torch.tensor([[1,2],[3,4]])
print("Shape of x:",x.shape)
print("Datatype of x:",x.dtype)
print(x.type)
```
