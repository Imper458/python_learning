```python

import torch
```

```python
#1.reshape:改变张量形状
x = torch.arange(1,13)
print("Original tensor:\n",x)
y = x.reshape(3,4)
print("Reshaped tensor (3x4):\n",y)
```

```python
#2.view：与reshape类似，但是要求张量是连续的
print("View to 4x3 tensor:\n",y.view(4,3))
```

```python
#3.transpose:转置张量
x = torch.randn(2,3,4)#创建2个3行4列的随机张量
print("Original tensor:\n",x)
y = x.transpose(0,2).shape#交换第0维和第2维，即将2个3行4列的张量变为4个3行2列的张量
print("Transposed tensor (dim1<->dim2):\n",y)
```

```python
#4.squeeze / unsqueeze:删除/增加维度
x = torch.randn(1,3,1,4)
y = x.squeeze()#删除所有维度为1的维度
print("Squeezed tensor shape:\n",y.shape)
z = y.unsqueeze(0)#在第0维增加一个维度
print("Unsqueezed tensor shape:\n",z.shape)
```

```python
#什么是连续张量？
#连续张量：张量的元素在内存中按照其索引顺序连续存储。
#非连续张量：张量的元素在内存中不是按照其索引顺序

```

```python
#连续张量的例子
#创建一个连续张量
a = torch.tensor([[1, 2, 3], [4, 5, 6]])
print("Is 'a' contiguous?:", a.is_contiguous())
#可以直接使用view方法改变形状
b = a.view(3, 2)
print("Reshaped tensor 'b' using view:\n", b)
```

```python
#非连续张量的例子
#通过转置创建一个非连续张量
a = torch.tensor([[1, 2, 3], [4, 5, 6]])
b = a.transpose(0, 1)
print("Is 'b' contiguous?:", b.is_contiguous())
#尝试使用view方法会报错
```

```python
#如果需要对非连续张量使用view方法，可以先调用.contiguous()方法将其转换为连续张量，然后再使用view方法。
c = b.contiguous().view(3, 2)
print("Reshaped tensor 'c' after making contiguous:\n", c)
```
