```python
import torch
```

```python
#1.张量的加法
a = torch.tensor([[1,2],[3,4]])
b = torch.tensor([[5,6],[7,8]])
c = a + b
print("Tensor addition:\n",c)
```

```python
#2.张量的乘法
d = a * b
print("Tensor multiplication:\n",d)
```

```python
#3.张量的减法
e = b - a
print("Tensor subtraction:\n",e)
```

```python
#4.张量矩阵乘法
f = torch.matmul(a,b)
print("Tensor matrix multiplication:\n",f)
```

```python
#5.张量逐个元素乘法
g = a * b
print("Tensor element-wise multiplication:\n",g)
```

```python
#6.张量拼接cat(dim=0 按照第一个维度、行拼接; dim=1 按照第二个维度、列拼接)
h = torch.cat([a,b],dim=0)
print("Tensor concatenation along dim=0:\n",h)
i = torch.cat([a,b],dim=1)
print("Tensor concatenation along dim=1:\n",i)
j = torch.cat([a,b],dim=-1)#-1表示最后一个维度
print("Tensor concatenation along dim=-1:\n",j)
```
